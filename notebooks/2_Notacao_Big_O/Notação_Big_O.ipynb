# Notação Big-O

- Como comparar dois algoritmos?
- Comparação objetiva entre algoritmos
- Considera diferenças entre poder de processamento, sistema operacional, linguagem de programação
- O quanto a "complexidade" do algoritmo aumenta de acordo com as entradas

## Função 1 - O(n)

```python
# 11 passos
def soma1(n):
    soma = 0
    for i in range(n + 1):
        soma += i

    return soma
```

```python
soma1(10)
```

```python
%timeit soma1(10)
```

## Função 2

```python
(10 * 11) / 2
```

```python
# 3 passos
# O(3)
def soma2(n):
    return (n * (n + 1)) / 2
```

```python
soma2(10)
```

```python
%timeit soma2(10)
```

## Função 3

- https://wiki.python.org/moin/TimeComplexity
- https://www.ics.uci.edu/~pattis/ICS-33/lectures/complexitypython.txt

```python
# O(1000) -> O(n)
def lista1():
    lista = []
    for i in range(1000):
        lista += [i]
    return lista
```

```python
print(lista1())
```

```python
len(lista1())
```

```python
%timeit lista1()
```

```python
labels = [
    "Constant",
    "Logarithmic",
    "Linear",
    "Log linear",
    "Quadratic",
    "Cubic",
    "Exponential",
]

big_o = [np.ones(n.shape), np.log(n), n, n * np.log(n), n**2, n**3, 2**n]
```

```python
plt.figure(figsize=(10, 8))  # tamanho da figura
plt.ylim(0, 100)  # limite do eixo y do gráfico
for i in range(len(big_o)):  # plotar todas as linhas num mesmo gráfico
    plt.plot(
        n, big_o[i], label=labels[i]
    )  # n para os dados,os valores big_o para tempo de execução e os labels para os nomes dos tipos de big-O
plt.legend()  # para termos legenda
plt.ylabel(
    "Runtime"
)  # os dados dos gráficos ficarão no eixo y. Runtime é tempo de execução.
plt.xlabel("n")  # n é o numero de entradas que passamos para o algoritmo.
```

## Exemplos Big-O

- https://www.bigocheatsheet.com/

### Constant - O(1)

```python
lista = [1, 2, 3, 4, 5]
```

```python
def constant(n):
    print(n[0])
```

```python
constant(lista)
```

### Linear - O(n)

```python
def linear(n):
    for i in n:
        print(i)
```

```python
linear(lista)
```

### Quadratic - O(n^2) - polynomial

```python
def quadratic(n):
    for i in n:
        for j in n:
            print(i, j)
        print("---")
```

```python
quadratic(lista)
```

### Combination

```python
# O(1) + O(5) + O(n) + O(n) + O(3)
# O(9) + O(2n) -> O(n)


def combination(n):
    # O(1)
    print(n[0])

    # O(5)
    for i in range(5):
        print("test ", i)

    # O(n)
    for i in n:
        print(i)

    # O(n)
    for i in n:
        print(i)

    # O(3)
    print("Python")
    print("Python")
    print("Python")
```

```python
combination(lista)
```
